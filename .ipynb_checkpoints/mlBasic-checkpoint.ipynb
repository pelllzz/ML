```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris_data = iris.data
iris_label = iris.target

print('iris target value: ', iris_label)
print('iris target name: ', iris.target_names)
```

```python
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df.head(3)
```

```python
iris_df.info()
```

```python
iris_df.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# random_state => 쉽게 생각하면 seed 랜덤으로 분류된 것들 중에 고정으로 쓰려고

X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size = 0.2, random_state=11)
```

```python
X_train.shape
```

```python
# DecisionTreeClassifier 객체 생성 및 학습 수행
# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html?highlight=decisiontreeclassifier#sklearn.tree.DecisionTreeClassifier
dt_clf = DecisionTreeClassifier(random_state=11)
dt_clf.fit(X_train, y_train)
```

```python
# 학습 완료 후 DecisionTreeClassifier 객체에서 테스트 셋으로 예측 수행
pred = dt_clf.predict(X_test)
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html#sklearn.metrics.accuracy_score
# 우리가 예측값이랑 실제값이랑 비교해도 좋은데 그걸... 손으로 해? 일해라 기계!!
from sklearn.metrics import accuracy_score
print('예측정확도: {0: .4f}'.format(accuracy_score(y_test,pred)))
```

```python
print(f'예측정확도:{accuracy_score(pred, y_test):.4f}')
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(10,7))
plot_tree(dt_clf)
plt.show()
```

```python
# filled - 클래스에 맞게 노드의 색을 칠함
# feature_names - 특성의 이름을 전달
# gini (gini impurity) - 불순도
plt.figure(figsize=(20,15))
plot_tree(dt_clf, filled=True, feature_names=iris.feature_names)
plt.show()
```
